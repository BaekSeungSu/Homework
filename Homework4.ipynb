3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

  - 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
  > range(stop)<br/>
  > range(start, stop[, step])<br/>
  - 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
  - 3) frange 사용 예
    - 3-1) frange(0.5)
      - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
      - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
      - [2.2, 2.7, 3.2, 3.7]

```python
def frange(start, stop=None, step=0.1):
    if stop == None: 
        stop = start
        start = 0
    ListOfFrange = []
    while start < stop:
        ListOfFrange.append(start)
        start += step
    return ["%.1f" % i for i in ListOfFrange]
print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2,4.0,0.5))
```

4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

  - sum() 사용 예

> sum()<br/>
>   0

> sum(1, 2)<br/>
>   3

> sum(1, 2, 3, 4, 5)<br/>
>   15

> sum(1, 5, 7, 2, -10)<br/>
>   5

```python
def sum(*args):
    sumOfargs = 0
    for i in args : sumOfargs += i
    return sumOfargs

print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))

```

```python
def upperString(s):
    return s[0].upper()
def myinitial(s):
    return "".join(map(upperString,s.split()))
print(myinitial("as soon as possible"))
print(myinitial("wow you are good!"))
```

6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
  - [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 

```python
def myfact(n):
    if n<=1: return 1
    else : return n * myfact(n-1)
print(myfact(2))
print(myfact(3))
print(myfact(30))
print(myfact(100))
print(myfact(1000))
print(myfact(2800))
```

  - 7) [주의] 필수사항
    - 위에서 만든 검색엔진 코딩은 매우 간단한 것이라 부족한 점이 많이 존재한다.
    - 본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 1가지 이상 제시하시오. 

```python
import os
import requests
import re
import string
import html
import pickle
import math
stopwords = []
with open("stopwords.txt", "r", encoding='utf-8') as f:
    for line in f: stopwords.append(line.strip()) 
        
        
def saveHTML(url):
    with open("%s.html" % url.split('//')[1].replace('/', '_'), "w", encoding='utf-8') as f:
        f.write(requests.get(url).text) 

def pickKey(urls):
    for u in urls:
        with open('%s.words.frequency' % u.split('//')[1].replace('/', '_'), "rb") as f:
            keywords = list(pickle.load(f).items()) 
            keywords.sort(key=lambda k: k[1], reverse=True)
            print("%s 에서 가장 많이 출현한 단어 3개 :" % u)
            for i in range(3):
                print("(%s, %d번)" % (keywords[i][0], keywords[i][1]), end=', ') #단어 3개 출력
            print()
            
def distance(d):
    return pow(math.e, -d) # y=e^(-|d|) (단, d는 문자와 문자 사이의 거리)

    
```
