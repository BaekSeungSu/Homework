    > \>\>\> c = c + 5  <br/>

    > [Count (step: 1)] 17 <br/>

  - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> print(c > 10)  <br/>

    > True <br/>
    > \>\>\> print(c == 17)  <br/>
    > True <br/> 
    > \>\>\> print(d != 9) <br/>
    > False <br/>

```python
class Counter:
    def __init__(self, count,step=1): #생성자
        self.count = count
        self. step = step
    def __str__(self): 
        return "[Count(step:{})]{}".format(self.step,self.count) #포멧팅사용
    def incr(self): 
        self.count+=self.step #count에 step값을 더함
    def __call__(self):
        self.incr()
    def __add__(self,other):
        return Counter(self.count + other,self.step)
    def __sub__(self,other):
        return Counter(self.count - other,self.step)
    def __gt__(self,other):
        return self.count > other
    def __lt__(self,other):
        return self.count < other
    def __eq__(self,other):
        return self.count == other
    def __ne__(self,other):
        return self.count != other

c = Counter(10)
d = Counter(10,2)
print(c)
print(d)
print('-' * 20)
c.incr()
d.incr()
print(c)
print(d)
print('-' * 20)
c()
d()
print(c)
print(d)
print('-' * 20)
c=c+5
d=d-5
print(c)
print(d)
print('-' * 20)
print(c>10)
print(d>10)
print(c<10)
print(d<10)
print(c==17)
print(d!=9)
```

  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

  - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> u = s | t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> u = s & t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {2, 3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> u = s - t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self,other):
        unionSet = self+other
        MySet.eliminate_duplicate(unionSet)
        return sorted(unionSet)

    def __and__(self,other):
        l=self+other
        interSet = []
        for i in l:
            if i in self and i in other:
                interSet.append(i)
        MySet.eliminate_duplicate(interSet)
        return sorted(interSet)
    
    def __sub__(self,other):
        differSet=[]
        for i in self:
            if i not in other:
                differSet.append(i)
        MySet.eliminate_duplicate(differSet)
        return sorted(differSet)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s|t
    i = s&t
    d = s-t
    print(u)
    print(i)
    print(d)
    
    
```

#### 문제 설명
1. 연산자 중복을 이용해서 __or__ , __and__ , __sub__ 메서드를 새롭게 작성한다.
2. __or__ 은 self와 other을 더하고 eliminate_duplicate를 통해 중복을 제거해서 합집합을 만들고 정렬시켜 반환한다.
3. __and__ 는 self와 other을 더한 l 리스트에 for문과 if문을 이용하여 l 리스트안에 값이 self와 other 둘다에 존재할경우 추가하도록하고 eliminate_duplicate를 통해 중복을 제거해서 교집합을 만들고 정렬시켜 반환한다.
4. __sub__ 는 for문과 if문을 통해 self의 원소가 other에 들어있지 않으면 differSet리스트에 추가하고 eliminate_duplicate 통해 중복을 제거해 차집합을 만들고 정렬시켜 반환한다.

```python
import requests
import re
import string
import html
class SearchEngine:
    stopwords = []
    #불용어 리스트 추가
    with open("stopwords.txt","r",encoding='utf-8') as f:
        for line in f:
            stopwords.append(line.strip())
            
    def __init__(self,*args): #생성자
        self.urlList = [] #주소를 저장할 리스트
        self.frequencyDictwithUrl = {} #주소 : 주소 : {키워드 : 개수}  형식으로 저장할 사전
        self.similarityDictwithUrl = {} #주소 : 유사도 형태 사전
        for i in args: #가변 인자로 주소를 받아
            self.addUrl(i) #addUrl 메서드를 이용해서 추가
        
    def addUrl(self,url): #주소 추가 메서드
        self.urlList.append(url) #주소 리스트에 추가
        self.frequencyDictwithUrl[url] = self.countWords(url) #해당 주소 단어 분석 결과 추가
        
        
    def removeUrl(self,url): #주소 제거 메서드
        if url in self.urlList: #주소리스트에 있는 경우에만
            self.urlList.remove(url) #주소리스트에서 제거
            del self.frequencyDictwithUrl[url] #사전에서 제거
            
    def listUrls(self): #전체 주소 출력 메서드
        for i in self.urlList: #주소리스트 각 요소 출력
            print(i)
            
    def getWordsFrequency(self): #단어 출현 빈도
        if not self.urlList: return None  # 등록된 url이 아니면 None 반환 
        frequencyDict = {}  #빈도수 사전 초기화
        for v in self.frequencyDictwithUrl.values(): #사전의 모든 value {키워드 : 개수}에 대해서 
            for key, value in v.items(): # 키워드 : 개수에 대해서
                if key in frequencyDict: frequencyDict[key] += value  # 단어가 이미있다면 개수 추가
                else: frequencyDict[key] = value #단어가 없으면 사전에 추가
        return frequencyDict  #사전 반환
    
    def getMaxFreqencyWords(self):# 최대 빈도
        if not self.urlList: return None  # 등록된 url이 아니면 None 반환 
        keywords = list(self.getWordsFrequency().items()) #빈도수 사전을 리스트로
        keywords.sort(key=lambda k: k[1], reverse=True) #k[1] 은 value, 즉 value를 기준으로 내림차순 정렬
        maxFrequency = keywords[0][1] #최대 빈도수 저장
        maxList = []
        for k in keywords: #리스트화된 사전에 대해서
            if k[1] == maxFrequency: maxList.append(k[0]) #최대빈도수가 동일하다면 추가
            else: break #그렇지 않을경우 break
        return maxList #최대 빈도수 리스트 반환
    
    def searchUrlByWord(self, keyword): #유사도 높은 웹사이트 출력 
        if not self.urlList: return None # 등록된 url이 아니라면 None 반환 
        for u in self.urlList:
            #유사도 분석
            self.similarityDictwithUrl[u] = self.similarityValidation(u, keyword) 
        sortedFrequency = list(self.similarityDictwithUrl.items())  #결과를 list화
        sortedFrequency.sort(key=lambda k: k[1], reverse=True) #value를 기준으로 내림차순정렬
        maxFrequency = sortedFrequency[0][1]# 최대 유사도 저장 
        maxList = []
        for s in sortedFrequency:
            if s[1] == maxFrequency: maxList.append(s[0]) #최대유사도가 동일하다면 추가
            else: break #그렇지 않을경우 break
        return maxList #최대 빈도수 리스트 반환 
    
    def countWords(self, url): #단어 빈도 저장
        req = requests.get(url) #사이트에서 html 문자열로 받아
        source = re.sub(r'\n', ' ', html.unescape(req.text)) #줄바꿈은 여백으로
        source = re.sub(r'<(script|style).*?</\1>', '', source) #script, style 제거
        source = re.sub(r'<!(--)?.*?-->', '', source) #주석제거
        source = re.sub(r'</?.*?>', '<>', source) #태그제거
        line = source.split('<>') #임의로 파싱
        words = []
        resultDic = {}
        for i in line: 
            if i.strip() == '': continue #공백일때
            result = [r for r in i.strip().split()] #단어사이 공백을 기준으로 파싱
            for r in range(len(result)):
                for p in string.punctuation:
                     result[r] = result[r].replace(p, '') #punctuation 제거
                if result[r].strip() == '': continue #punctuation제거했는데 공백일경우
                if result[r] in resultDic:
                    resultDic[result[r]] += 1  # 단어가 저장되어 있는경우 +1
                else: 
                    if result[r] not in self.stopwords: resultDic[result[r]] = 1 #불용어 제거 및 단어개수 저장
        return resultDic
                        
    
```

#### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 ***오버라이드***하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
  - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
  - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
    - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
    
  > \>\>\> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
    
  > \>\>\> w4.getWordsFrequency()<br/>
  > [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
    
  > \>\>\> w4.getWordsFrequency(reverse=true)<br/>
  > [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine): #SearchEngine을 상속으로 하는 클래스 선언
    def __iter__(self): #반복자
        return iter(self.global_keywords) #리스트를 반복자로 반환
    
    def getWordsFrequency(self, reverse=False):
        if not self.urlList: return None #등록된 주소가 없으면 None 객체 반환
        frequencyDict = {} #단어 분석 결과들을 병합할 사전 생성
        for v in self.frequencyDictwithUrl.values():
            for key, value in v.items():
                if key in frequencyDict: frequencyDict[key] += value #단어가 이미 있다면 개수만큼 더함
                else: frequencyDict[key] = value #없다면 추가함
        self.global_keywords = list(frequencyDict.items()) #병합된 빈도수 사전을 리스트로 변환 후
        self.global_keywords.sort(key=lambda k: k[1], reverse=not reverse) #정렬
        self.global_keywords = list(self.global_keywords) #리스트로 변환
        return self.global_keywords #정렬된 리스트 반환
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
```
