```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self,other):
        unionSet = self+other
        MySet.eliminate_duplicate(unionSet)
        return sorted(unionSet)

    def __and__(self,other):
        l=self+other
        interSet = []
        for i in l:
            if i in self and i in other:
                interSet.append(i)
        MySet.eliminate_duplicate(interSet)
        return sorted(interSet)
    
    def __sub__(self,other):
        differSet=[]
        for i in self:
            if i not in other:
                differSet.append(i)
        MySet.eliminate_duplicate(differSet)
        return sorted(differSet)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s|t
    i = s&t
    d = s-t
    print(u)
    print(i)
    print(d)
    
    
```

```python
import requests
import re
import string
import html
class SearchEngine:
    stopwords = []
    #ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ ì¶”ê°€
    with open("stopwords.txt","r",encoding='utf-8') as f:
        for line in f:
            stopwords.append(line.strip())
            
    def __init__(self,*args): #ìƒì„±ì
        self.urlList = [] #ì£¼ì†Œë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
        self.frequencyDictwithUrl = {} #ì£¼ì†Œ : ì£¼ì†Œ : {í‚¤ì›Œë“œ : ê°œìˆ˜}  í˜•ì‹ìœ¼ë¡œ ì €ì¥í•  ì‚¬ì „
        self.similarityDictwithUrl = {} #ì£¼ì†Œ : ìœ ì‚¬ë„ í˜•íƒœ ì‚¬ì „
        for i in args: #ê°€ë³€ ì¸ìë¡œ ì£¼ì†Œë¥¼ ë°›ì•„
            self.addUrl(i) #addUrl ë©”ì„œë“œë¥¼ ì´ìš©í•´ì„œ ì¶”ê°€
        
    def addUrl(self,url): #ì£¼ì†Œ ì¶”ê°€ ë©”ì„œë“œ
        self.urlList.append(url) #ì£¼ì†Œ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        self.frequencyDictwithUrl[url] = self.countWords(url) #í•´ë‹¹ ì£¼ì†Œ ë‹¨ì–´ ë¶„ì„ ê²°ê³¼ ì¶”ê°€
        
        
    def removeUrl(self,url): #ì£¼ì†Œ ì œê±° ë©”ì„œë“œ
        if url in self.urlList: #ì£¼ì†Œë¦¬ìŠ¤íŠ¸ì— ìˆëŠ” ê²½ìš°ì—ë§Œ
            self.urlList.remove(url) #ì£¼ì†Œë¦¬ìŠ¤íŠ¸ì—ì„œ ì œê±°
            del self.frequencyDictwithUrl[url] #ì‚¬ì „ì—ì„œ ì œê±°
            
    def listUrls(self): #ì „ì²´ ì£¼ì†Œ ì¶œë ¥ ë©”ì„œë“œ
        for i in self.urlList: #ì£¼ì†Œë¦¬ìŠ¤íŠ¸ ê° ìš”ì†Œ ì¶œë ¥
            print(i)
            
    def getWordsFrequency(self): #ë‹¨ì–´ ì¶œí˜„ ë¹ˆë„
        if not self.urlList: return None  # ë“±ë¡ëœ urlì´ ì•„ë‹ˆë©´ None ë°˜í™˜ 
        frequencyDict = {}  #ë¹ˆë„ìˆ˜ ì‚¬ì „ ì´ˆê¸°í™”
        for v in self.frequencyDictwithUrl.values(): #ì‚¬ì „ì˜ ëª¨ë“  value {í‚¤ì›Œë“œ : ê°œìˆ˜}ì— ëŒ€í•´ì„œ 
            for key, value in v.items(): # í‚¤ì›Œë“œ : ê°œìˆ˜ì— ëŒ€í•´ì„œ
                if key in frequencyDict: frequencyDict[key] += value  # ë‹¨ì–´ê°€ ì´ë¯¸ìˆë‹¤ë©´ ê°œìˆ˜ ì¶”ê°€
                else: frequencyDict[key] = value #ë‹¨ì–´ê°€ ì—†ìœ¼ë©´ ì‚¬ì „ì— ì¶”ê°€
        return frequencyDict  #ì‚¬ì „ ë°˜í™˜
    
    def getMaxFreqencyWords(self):# ìµœëŒ€ ë¹ˆë„
        if not self.urlList: return None  # ë“±ë¡ëœ urlì´ ì•„ë‹ˆë©´ None ë°˜í™˜ 
        keywords = list(self.getWordsFrequency().items()) #ë¹ˆë„ìˆ˜ ì‚¬ì „ì„ ë¦¬ìŠ¤íŠ¸ë¡œ
        keywords.sort(key=lambda k: k[1], reverse=True) #k[1] ì€ value, ì¦‰ valueë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
        maxFrequency = keywords[0][1] #ìµœëŒ€ ë¹ˆë„ìˆ˜ ì €ì¥
        maxList = []
        for k in keywords: #ë¦¬ìŠ¤íŠ¸í™”ëœ ì‚¬ì „ì— ëŒ€í•´ì„œ
            if k[1] == maxFrequency: maxList.append(k[0]) #ìµœëŒ€ë¹ˆë„ìˆ˜ê°€ ë™ì¼í•˜ë‹¤ë©´ ì¶”ê°€
            else: break #ê·¸ë ‡ì§€ ì•Šì„ê²½ìš° break
        return maxList #ìµœëŒ€ ë¹ˆë„ìˆ˜ ë¦¬ìŠ¤íŠ¸ ë°˜í™˜
    
    def searchUrlByWord(self, keyword): #ìœ ì‚¬ë„ ë†’ì€ ì›¹ì‚¬ì´íŠ¸ ì¶œë ¥ 
        if not self.urlList: return None # ë“±ë¡ëœ urlì´ ì•„ë‹ˆë¼ë©´ None ë°˜í™˜ 
        for u in self.urlList:
            #ìœ ì‚¬ë„ ë¶„ì„
            self.similarityDictwithUrl[u] = self.similarityValidation(u, keyword) 
        sortedFrequency = list(self.similarityDictwithUrl.items())  #ê²°ê³¼ë¥¼ listí™”
        sortedFrequency.sort(key=lambda k: k[1], reverse=True) #valueë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœì •ë ¬
        maxFrequency = sortedFrequency[0][1]# ìµœëŒ€ ìœ ì‚¬ë„ ì €ì¥ 
        maxList = []
        for s in sortedFrequency:
            if s[1] == maxFrequency: maxList.append(s[0]) #ìµœëŒ€ìœ ì‚¬ë„ê°€ ë™ì¼í•˜ë‹¤ë©´ ì¶”ê°€
            else: break #ê·¸ë ‡ì§€ ì•Šì„ê²½ìš° break
        return maxList #ìµœëŒ€ ë¹ˆë„ìˆ˜ ë¦¬ìŠ¤íŠ¸ ë°˜í™˜ 
    
    def countWords(self, url): #ë‹¨ì–´ ë¹ˆë„ ì €ì¥
        req = requests.get(url) #ì‚¬ì´íŠ¸ì—ì„œ html ë¬¸ìì—´ë¡œ ë°›ì•„
        source = re.sub(r'\n', ' ', html.unescape(req.text)) #ì¤„ë°”ê¿ˆì€ ì—¬ë°±ìœ¼ë¡œ
        source = re.sub(r'<(script|style).*?</\1>', '', source) #script, style ì œê±°
        source = re.sub(r'<!(--)?.*?-->', '', source) #ì£¼ì„ì œê±°
        source = re.sub(r'</?.*?>', '<>', source) #íƒœê·¸ì œê±°
        line = source.split('<>') #ì„ì˜ë¡œ íŒŒì‹±
        words = []
        resultDic = {}
        for i in line: 
            if i.strip() == '': continue #ê³µë°±ì¼ë•Œ
            result = [r for r in i.strip().split()] #ë‹¨ì–´ì‚¬ì´ ê³µë°±ì„ ê¸°ì¤€ìœ¼ë¡œ íŒŒì‹±
            for r in range(len(result)):
                for p in string.punctuation:
                     result[r] = result[r].replace(p, '') #punctuation ì œê±°
                if result[r].strip() == '': continue #punctuationì œê±°í–ˆëŠ”ë° ê³µë°±ì¼ê²½ìš°
                if result[r] in resultDic:
                    resultDic[result[r]] += 1  # ë‹¨ì–´ê°€ ì €ì¥ë˜ì–´ ìˆëŠ”ê²½ìš° +1
                else: 
                    if result[r] not in self.stopwords: resultDic[result[r]] = 1 #ë¶ˆìš©ì–´ ì œê±° ë° ë‹¨ì–´ê°œìˆ˜ ì €ì¥
        return resultDic
                        
    
```

#### 2. ìœ„ 1ë²ˆ ë¬¸ì œì—ì„œ ì •ì˜í•œ SearchEngine í´ë˜ìŠ¤ë¥¼ ìƒì†í•˜ì—¬ SearchEngineWithOrderedWebWords í´ë˜ìŠ¤ë¥¼ ì •ì˜í•˜ê³  ìŠˆí¼í´ë˜ìŠ¤ì— ì •ì˜ëœ getWordsFrequency() ë©”ì†Œë“œë¥¼ ***ì˜¤ë²„ë¼ì´ë“œ***í•˜ì—¬ ë‹¨ì–´ ì¶œí˜„ ë¹ˆë„ë¥¼ ë‚´ë¦¼ ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ì¶œë ¥í•˜ì‹œì˜¤.

  - getWordFrequence() ë©”ì†Œë“œì— reverse ë¼ëŠ” ì¸ìë¥¼ ë§Œë“¤ê³  true ë˜ëŠ” falseë¥¼ ì¸ìë¡œ ë°›ì„ ìˆ˜ ìˆë„ë¡ í•œë‹¤.
    - reverse ì¸ìì˜ ë””í´íŠ¸ ê°’ì€ false ì´ë©°, ê¸°ë³¸ì ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤.
    - reverse ì¸ìì— trueë¥¼ ë„£ìœ¼ë©´ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤.
    
  > \>\>\> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
    
  > \>\>\> w4.getWordsFrequency()<br/>
  > [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
    
  > \>\>\> w4.getWordsFrequency(reverse=true)<br/>
  > [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine): #SearchEngineì„ ìƒì†ìœ¼ë¡œ í•˜ëŠ” í´ë˜ìŠ¤ ì„ ì–¸
    def __iter__(self): #ë°˜ë³µì
        return iter(self.global_keywords) #ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜ë³µìë¡œ ë°˜í™˜
    
    def getWordsFrequency(self, reverse=False):
        if not self.urlList: return None #ë“±ë¡ëœ ì£¼ì†Œê°€ ì—†ìœ¼ë©´ None ê°ì²´ ë°˜í™˜
        frequencyDict = {} #ë‹¨ì–´ ë¶„ì„ ê²°ê³¼ë“¤ì„ ë³‘í•©í•  ì‚¬ì „ ìƒì„±
        for v in self.frequencyDictwithUrl.values():
            for key, value in v.items():
                if key in frequencyDict: frequencyDict[key] += value #ë‹¨ì–´ê°€ ì´ë¯¸ ìˆë‹¤ë©´ ê°œìˆ˜ë§Œí¼ ë”í•¨
                else: frequencyDict[key] = value #ì—†ë‹¤ë©´ ì¶”ê°€í•¨
        self.global_keywords = list(frequencyDict.items()) #ë³‘í•©ëœ ë¹ˆë„ìˆ˜ ì‚¬ì „ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜ í›„
        self.global_keywords.sort(key=lambda k: k[1], reverse=not reverse) #ì •ë ¬
        self.global_keywords = list(self.global_keywords) #ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
        return self.global_keywords #ì •ë ¬ëœ ë¦¬ìŠ¤íŠ¸ ë°˜í™˜
```
